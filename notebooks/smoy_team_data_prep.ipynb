```python
# Import Team Data
import pandas as pd
team_data_df = pd.read_csv('../unprocessed-data/team-data/nba_team_win_totals.csv')
# Check if load correctly
pd.set_option('display.max_columns', None)
team_data_df
```

```python
# Modify 'Season' values from '2023-2024' to '2023
team_data_df['Season'] = team_data_df['Season'].str.split('-').str[0]
# Drop Lg (League) and Rk (Rank) columns
team_data_df = team_data_df.drop(columns=['Lg', 'Rk'])
# Drop all rows with column 'Season' < 1980 (player data only up until then) and 'Season' = Total (needs recalculation)
team_data_df = team_data_df[team_data_df['Season'] != 'Total']
team_data_df = team_data_df[team_data_df['Season'].astype(int) >= 1980]
# Check
team_data_df.head()
```

```python
# Remove unnecessary decimals (Wins can only be whole numbers)
team_data_df = team_data_df.apply(pd.to_numeric, errors='ignore').astype('Int64', errors='ignore')

# Check
team_data_df
 
```

```python
# Write the prepared team data to a CSV file
file_path = '../processed-data/prepared_team_win_data.csv'
team_data_df.to_csv(file_path, index=False)
print("Data written to: ", file_path)
```
